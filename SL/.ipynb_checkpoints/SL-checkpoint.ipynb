```python
# Create binary target
cancer_df['Long_Term_Survival'] = (cancer_df['Survival_Years'] > 5).astype(int)

# Drop columns that leak or aren't helpful
cancer_features = cancer_df.drop(columns=['Patient_ID', 'Survival_Years', 'Target_Severity_Score', 'Long_Term_Survival'])
cancer_target = cancer_df['Long_Term_Survival']
```

```python
from sklearn.preprocessing import LabelEncoder, StandardScaler
from sklearn.model_selection import train_test_split

# Encode categorical features
cancer_encoded = cancer_features.copy()
cat_cols = cancer_encoded.select_dtypes(include='object').columns
le = LabelEncoder()
for col in cat_cols:
    cancer_encoded[col] = le.fit_transform(cancer_encoded[col])

# Standard scale
scaler = StandardScaler()
cancer_scaled = scaler.fit_transform(cancer_encoded)

# Split
X_cancer_train, X_cancer_test, y_cancer_train, y_cancer_test = train_test_split(
    cancer_scaled, cancer_target, test_size=0.2, random_state=42, stratify=cancer_target)
```

```python
# Clean column names
bankrupt_df.columns = bankrupt_df.columns.str.strip()

# Drop NaNs (if needed)
bankrupt_df = bankrupt_df.dropna()

# Features and target
X_bankrupt = bankrupt_df.drop(columns=['Bankrupt?'])
y_bankrupt = bankrupt_df['Bankrupt?']
```

```python
# Standard scale
X_bankrupt_scaled = scaler.fit_transform(X_bankrupt)

# Split
X_bankrupt_train, X_bankrupt_test, y_bankrupt_train, y_bankrupt_test = train_test_split(
    X_bankrupt_scaled, y_bankrupt, test_size=0.2, random_state=42, stratify=y_bankrupt)
```

# Modeling functions

```python
from sklearn.model_selection import learning_curve, validation_curve
import numpy as np

def plot_learning_curve(estimator, X, y, title):
    train_sizes, train_scores, test_scores = learning_curve(estimator, X, y,
                                                            cv=5, scoring='accuracy',
                                                            train_sizes=np.linspace(0.1, 1.0, 5),
                                                            n_jobs=-1)
    train_mean = train_scores.mean(axis=1)
    test_mean = test_scores.mean(axis=1)

    plt.plot(train_sizes, train_mean, label='Training accuracy')
    plt.plot(train_sizes, test_mean, label='Testing accuracy')
    plt.title(f'Learning Curve: {title}')
    plt.xlabel('Training Size')
    plt.ylabel('Accuracy')
    plt.legend()
    plt.grid()
    plt.show()
```

# Algorithm KNN

```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score

# Try k=3, 5, 11
for k in [3, 5, 11]:
    knn = KNeighborsClassifier(n_neighbors=k)
    knn.fit(X_cancer_train, y_cancer_train)
    print(f"k={k} | Train: {accuracy_score(y_cancer_train, knn.predict(X_cancer_train)):.3f} | Test: {accuracy_score(y_cancer_test, knn.predict(X_cancer_test)):.3f}")

# Plot learning curve
plot_learning_curve(KNeighborsClassifier(n_neighbors=5), X_cancer_train, y_cancer_train, "k-NN (Cancer)")
```

```python
for k in [3, 5, 11]:
    knn = KNeighborsClassifier(n_neighbors=k)
    knn.fit(X_bankrupt_train, y_bankrupt_train)
    print(f"k={k} | Train: {accuracy_score(y_bankrupt_train, knn.predict(X_bankrupt_train)):.3f} | Test: {accuracy_score(y_bankrupt_test, knn.predict(X_bankrupt_test)):.3f}")

plot_learning_curve(KNeighborsClassifier(n_neighbors=5), X_bankrupt_train, y_bankrupt_train, "k-NN (Bankruptcy)")
```

# Algorithm SVM

```python
from sklearn.svm import SVC

# RBF kernel
svc_rbf = SVC(kernel='rbf', gamma='scale')
svc_rbf.fit(X_cancer_train, y_cancer_train)
print("SVM-RBF (Cancer) Test Accuracy:", accuracy_score(y_cancer_test, svc_rbf.predict(X_cancer_test)))
plot_learning_curve(svc_rbf, X_cancer_train, y_cancer_train, "SVM-RBF (Cancer)")

# Linear kernel
svc_lin = SVC(kernel='linear')
svc_lin.fit(X_bankrupt_train, y_bankrupt_train)
print("SVM-Linear (Bankrupt) Test Accuracy:", accuracy_score(y_bankrupt_test, svc_lin.predict(X_bankrupt_test)))
plot_learning_curve(svc_lin, X_bankrupt_train, y_bankrupt_train, "SVM-Linear (Bankruptcy)")
```

# Algorithm NN (MLP)

```python
from sklearn.neural_network import MLPClassifier

# Cancer Dataset
mlp_relu = MLPClassifier(hidden_layer_sizes=(50,), activation='relu', max_iter=300, random_state=1)
mlp_relu.fit(X_cancer_train, y_cancer_train)
print("NN (ReLU, Cancer) Test Accuracy:", accuracy_score(y_cancer_test, mlp_relu.predict(X_cancer_test)))
plot_learning_curve(mlp_relu, X_cancer_train, y_cancer_train, "Neural Net (ReLU, Cancer)")

mlp_tanh = MLPClassifier(hidden_layer_sizes=(50,), activation='tanh', max_iter=300, random_state=1)
mlp_tanh.fit(X_bankrupt_train, y_bankrupt_train)
print("NN (Tanh, Bankrupt) Test Accuracy:", accuracy_score(y_bankrupt_test, mlp_tanh.predict(X_bankrupt_test)))
plot_learning_curve(mlp_tanh, X_bankrupt_train, y_bankrupt_train, "Neural Net (Tanh, Bankruptcy)")
```

# Hyperparameter Validation Curves

```python
# KNN - Varying KNN
```

```python
# Validation curve for k in k-NN
from sklearn.model_selection import validation_curve

def plot_validation_curve_knn(X, y, title):
    param_range = np.arange(1, 31)
    train_scores, test_scores = validation_curve(
        KNeighborsClassifier(), X, y,
        param_name="n_neighbors",
        param_range=param_range,
        cv=5, scoring="accuracy", n_jobs=-1)

    plt.plot(param_range, train_scores.mean(axis=1), label="Train")
    plt.plot(param_range, test_scores.mean(axis=1), label="Test")
    plt.title(f"Validation Curve: {title}")
    plt.xlabel("k (n_neighbors)")
    plt.ylabel("Accuracy")
    plt.legend()
    plt.grid()
    plt.show()

plot_validation_curve_knn(X_cancer_train, y_cancer_train, "k-NN (Cancer)")
plot_validation_curve_knn(X_bankrupt_train, y_bankrupt_train, "k-NN (Bankruptcy)")
```

```python
# SVM - Varying C and Kernel 
```

```python
def plot_validation_curve_svm(X, y, kernel_type, title):
    param_range = np.logspace(-2, 2, 10)
    train_scores, test_scores = validation_curve(
        SVC(kernel=kernel_type), X, y,
        param_name="C",
        param_range=param_range,
        cv=5, scoring="accuracy", n_jobs=-1)

    plt.semilogx(param_range, train_scores.mean(axis=1), label="Train")
    plt.semilogx(param_range, test_scores.mean(axis=1), label="Test")
    plt.title(f"Validation Curve: {title}")
    plt.xlabel("C")
    plt.ylabel("Accuracy")
    plt.legend()
    plt.grid()
    plt.show()

plot_validation_curve_svm(X_cancer_train, y_cancer_train, "rbf", "SVM-RBF (Cancer)")
plot_validation_curve_svm(X_bankrupt_train, y_bankrupt_train, "linear", "SVM-Linear (Bankruptcy)")
```

```python
# 🔧 6C. Neural Net — Varying Alpha (Regularization Strength)
```

```python
def plot_validation_curve_nn(X, y, activation, title):
    param_range = np.logspace(-4, 1, 6)
    train_scores, test_scores = validation_curve(
        MLPClassifier(activation=activation, max_iter=500, random_state=1),
        X, y,
        param_name="alpha",
        param_range=param_range,
        cv=5, scoring="accuracy", n_jobs=-1)

    plt.semilogx(param_range, train_scores.mean(axis=1), label="Train")
    plt.semilogx(param_range, test_scores.mean(axis=1), label="Test")
    plt.title(f"Validation Curve: {title}")
    plt.xlabel("Alpha (L2 Penalty)")
    plt.ylabel("Accuracy")
    plt.legend()
    plt.grid()
    plt.show()

plot_validation_curve_nn(X_cancer_train, y_cancer_train, "relu", "NN-ReLU (Cancer)")
plot_validation_curve_nn(X_bankrupt_train, y_bankrupt_train, "tanh", "NN-Tanh (Bankruptcy)")
```
